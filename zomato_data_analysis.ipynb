```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("/content/Zomato data .csv")
print (dataframe)
```

```python
dataframe
```

```python
print(dataframe.info())
```

#Type of restaurents

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of Restaurant')
plt.ylabel('Count')
```

#conclusion:Majority of restuarants fall in type of dinning

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="b", marker="s")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
plt.grid(True)
```

#conclusion:majority of restuarants are having dinning as maximum vote


```python
plt.hist(dataframe['rate'], bins=4, edgecolor='blue', rwidth=0.5
         )
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
```

# Majority Restaurents have 3.5 to 4 ratings

#Avg order spending by couple  

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
plt.grid(True)
```

```python
dataframe.head()
```

```python
plt.figure(figsize=(5, 8))
sns.boxplot(x='online_order', y='rate', data=dataframe)
plt.grid(True)
```

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
